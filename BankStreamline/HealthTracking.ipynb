```python
robin_hood = 696
rrcu = 15674
pibank = 3520
beehive = 815
total = robin_hood + rrcu + pibank + beehive

print(f'{robin_hood + rrcu + pibank + beehive:,}')
```

#### Dates

```python
last_month = time.localtime().tm_mon - 1

current_yr = time.localtime().tm_year

# handling if last month was January
if last_month == 0:
    last_month = 12
    current_yr = current_yr - 1


first_of_month = str(current_yr) + '-' + str(last_month) + '-01'

print(last_month)
print(current_yr)
print(first_of_month)
```
